Vamos a utilizar el paquete Deepforest https://github.com/weecology/DeepForest para identificar aves a partir de imagenes aereas

# Datos de aves

```python
from deepforest import main
from deepforest import visualize
import matplotlib.pyplot as plt

# importa el modelo pre entrenado
m = main.deepforest()
# Para deteccion de aves
m.load_model("Weecology/deepforest-bird")

# Define la imagen para predecir
image_path = "example_airborne_birds/5.jpg"

# Hace la prediccion
predictions = m.predict_tile(image_path, patch_size=600, patch_overlap=0)
# visualiza los documentos
visualize.plot_results(predictions)
plt.show()
```

```python
# Evaluar las predicciones
predictions.score.hist()
```

```python
# Filtrar las mejores predicciones
high_confidence = predictions[predictions.score > 0.2]
print(f'las predicciones totales son {predictions.shape[0]}')
print(f'las predicciones de alta confianza son {high_confidence.shape[0]}')
```

```python
high_confidence.root_dir = "example_airborne_birds/"
visualize.plot_results(high_confidence)
```
